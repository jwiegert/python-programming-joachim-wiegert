```python
# start with any positive integer n. Then each term is obtained from the previous 
# term as follows: if the previous term is even, the next term is one half of the 
# previous term. If the previous term is odd, the next term is 3 times the previous 
# term plus 1. The conjecture is that no matter what value of n, the sequence will 
# always reach 1. 
import random as rnd
import matplotlib.pyplot as plt

Nsims = 10

for _ in range(Nsims):
    number = []
    N = rnd.randint(0,1000)
    niteration = 0
    print(f"Initial number: {N}")

    while N != 1:
        number.append(N)
        if N%2 == 0:
            N /= 2
        else:
            N = 3*N + 1
        niteration += 1
    print(f"Number iterations: {niteration}")
    plt.plot(number,'-.')
    plt.plot(len(number)-1,number[-1],'x',markersize = 10)
plt.xlabel("Number iterations")


```

```python
import matplotlib.pyplot as plt


#Ninitial = list(range(2,1000)) # Linear scale
Ninitial = [10**n for n in range(261)] # Log scale
Niterations = []

for N in Ninitial:
    niteration = 0
    while N != 1:
        if N%2 == 0:
            N = int(N/2)
        else:
            N = int(3*N + 1)
        niteration += 1
    Niterations.append(niteration)

fig,ax = plt.figure(dpi=100), plt.axes(xlabel="Initial number",ylabel="Number iterations", xscale='log');
plt.plot(Ninitial,Niterations);
```
