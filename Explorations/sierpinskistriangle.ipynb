```python
import random as rnd
import matplotlib.pyplot as plt

x = []
y = []

base = 10
height = 10
Niterations = 10000

p = [[0,0],[base,0],[0.5*base,height]]
k = height/(0.5*base)
m = height + k*0.5*base

print(k,m)

# Initial point, radnom inside the triangle
# X between 0 and 10
# Y below the two straight lines that borders the triangle
x.append(rnd.uniform(0,10))
if x[0] <= 0.5*base:
    y.append(rnd.uniform(0,k*x[0]))
else:
    y.append(rnd.uniform(0,m - k*x[0]))

for i in range(Niterations):
    # Choose a random edgepoint
    nedge = rnd.randint(0,2)

    # Put new point at half the distance between previous point and edge
    x.append(0.5*abs(p[nedge][0] - x[i]) + min(p[nedge][0],x[i]))
    y.append(0.5*abs(p[nedge][1] - y[i]) + min(p[nedge][1],y[i]))

plt.figure(dpi=100)
plt.plot(x,y,'b.',markersize=1)
plt.plot(x[0],y[0],'r.')


```

```python
x = list(range(0,11))
y1 = [k*xn for xn in x]
y2 = [-k*xn + m for xn in x]

plt.plot(x,y1)
plt.plot(x,y2)
```
