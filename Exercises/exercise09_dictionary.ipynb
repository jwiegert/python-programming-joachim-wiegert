# 2. Dice simulation (*)

Simulate 1000000 dice rolls and save the number of ones, twos, ..., sixes in a dictionary. Then print them out in the terminal. (*)

```python
import random as rnd

Nrolls = int(1e6)
diceresults = list(range(1,7))
textresults = ["Ones","Twos","Threes","Fours","Fives","Sixes"]
dicerolls   = {roll:0 for roll in textresults}

for nroll in range(Nrolls):
    for i in range(6):
        if rnd.randint(0,5) == i: # Note that 0 corresponds to One, etc to 5 corresponds to Six.
            dicerolls[textresults[i]] += 1
#
# Print nicely
#
for roll,nroll in dicerolls.items():
    print(f"{roll}: {nroll}")
```

# 3. Pokemon list (*)

Read in the file pokemon_list.txt in Python. Create a variable with name pokedex with the key:value "pokemon":"type, index". For example when searching for the keys "Gengar" and "Pikachu": (*)

```python
path = '../Code-Along/8_files/'
#
# Create empty dictionary
#
pokedex = {}
#
# Open file
#
with open(path+'pokemon_list.txt', 'r') as pokemonfile:
    #
    # Extract data
    #
    for data in pokemonfile.readlines():
        pokeindex = data.strip("\n").split()[0]
        pokemon   = data.strip("\n").split()[1]
        poketype  = data.strip("\n").split()[2]
        #
        # Add to dictionary
        #
        pokedex[pokemon] = f"{poketype}, {pokeindex}"
#
# Print a few
#
print(f"Pikachu: {pokedex['Pikachu']}")
print(f"Charizard: {pokedex['Charizard']}\n\n")
#
# Print the whole dictionary nicely
#
for pokemon,poketype in pokedex.items():
    print(f"{pokemon}: {poketype}")
```

# 4. Morse code (**)

Read in the file morse.txt, save it in a dictionary and create a function that lets the user input a message to get it translated to morse code. For example (**)

```python
import re
path = '../Code-Along/8_files/'
#
# Create empty dictionary
#
morsedict = {}
#
# Open file
#
with open(path+'morse.txt', 'r') as morsefile:
    #
    # Extract data
    #
    for data in morsefile.readlines():
        alphaletter = data.strip("\n").split()[0].replace(":", "")
        morseletter = data.strip("\n").split()[1]
        #
        # Add to dictionary
        #
        morsedict[alphaletter] = morseletter
#
# Function to translate text message to morse message
#
def writemorsemess(message):
    morsemessage = []                     # Place holder morsemessage list
    message = message.upper()             # Set all letters to upper case
    message = message.split()             # Separate words in different elements
    for word in message:                  # Loop through the words
        word = re.sub("\W", "", word)     # Remove all non-characters (\W)
        word = re.sub("\d", "", word)     # Remove all digits (\W)
        morseword = []                    # Create empty string for each word in morse
        for letter in word:               # Loop through letters into morse dictionary
            morseword.append(morsedict[letter]) # Save morse word
        morseword = "".join(morseword)    # Put together the morse letters
        morsemessage.append(morseword)    # Save the whole morse message
    morsemessage = " ".join(morsemessage) # Put together to one message (with spaces between words)
    return morsemessage
#
# Ask for a message
#
message = input("Write a message you want to translate to morse code (no numbers, only letters): ")
print(message)
print(writemorsemess(message))
```
