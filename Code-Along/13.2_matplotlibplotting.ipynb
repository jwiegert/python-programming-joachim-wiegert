# Matplotlib

- Matlab style
- OOP style

```python
import matplotlib.pyplot as plt
import numpy as np

# Create an x-array and a corresponding f(x) array
x = np.linspace(-5,5)
f = np.sin(x)
```

### MATLAB style interface

```python
plt.figure(dpi = 80)
plt.plot(x,f)
plt.title("sin(x)")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.figure(dpi = 80, figsize=(10,3))
plt.subplot(1,2,1) #row,columns,panelnumber
plt.plot(x,np.sin(x))
plt.plot(x,x**2/10-1)

plt.subplot(1,2,2)
plt.plot(x,np.cos(x))

# plt.show() if you're in a python script. STops the program so use this in the end
# Then you get something interactive
```

### OOP style

```python
# fig is a kind of object even though figure is a function.
# plt calls both classes and functions
fig = plt.figure(dpi=80)
# ax object can have a plot, axis, labels, title etc
ax = plt.axes()
# so we may now use ax.plot instead of plt.plot
ax.plot(x, f)
# And now we can set all the settings with attributes with set instead
ax.set(xlabel="x", ylabel="y", title=r"$\sin{x}$", xlim=(-4,4)); #; removes a list of settings in output
```

```python
# subplot here, subplot returns a tuple with two attributes, fig and ax.
fig, ax = plt.subplots(1,2, dpi=100, figsize=(10,3))
# Example, set up a list of normal distributed random numbers
rng = np.random.RandomState(42)
data_normal_dist = rng.randn(1000)

# Call first subplot, ax[0], ax[1] calls next figure, etc
# hist for histogram
# bins, number of buns
# alpga for transparency
ax[0].hist(data_normal_dist, bins=30, alpha=0.6);
ax[0].set(title="Histogram of normal distribution data", xlabel="x", ylabel="y")

#ax[1].plot(data_normal_dist,'.');

# Scatter plots
# This is actually the idea behind Hans Roslings plots
# x, y for data
# c is colour
# alpha is transperancy
# cmap is a colour map
# s is the size of each point
# Contraty to plot you can change the props of each point instead of all at once as you do with plot.
bubble = ax[1].scatter(x=rng.randn(100),y=rng.randn(100), c=rng.random(100), alpha=0.7, cmap="cividis", s=100*rng.randn(100));
ax[1].set(title="Bubble chart", xlabel="x", ylabel="y")
plt.savefig("randombubbleplot.png", facecolor="white")
```
